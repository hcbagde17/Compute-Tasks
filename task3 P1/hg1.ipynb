```python
import torch
from transformers import pipeline
```

```python
# initailizing the the chatbot pipeline
chatbot = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-0.5B-Instruct",
)
```

```python
# System Prompt
# keeping it simple for smaller models so that it doesn't get confused.
system_prompt = "You are a helpful AI assistant. Answer concisely."

# Chat History
messages = [
    {"role": "system", "content": system_prompt}
]
```

```python
# Chat Loop
while True:
    user_input = input("You: ")
    
    if user_input.lower() in ["exit", "quit"]:
        break
    print(f"you: {user_input}")
    messages.append({"role": "user", "content": user_input})

    # Generate
    outputs = chatbot(
        messages, 
        max_new_tokens=256, 
        do_sample=True, 
        temperature=0.7
    )
    
    reply = outputs[0]["generated_text"][-1]["content"]
    
    print(f"Bot: {reply}\n")
    
    messages.append({"role": "assistant", "content": reply})
```
