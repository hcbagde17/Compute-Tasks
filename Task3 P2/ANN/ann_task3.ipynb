```python
# imputing missing values in 'Popularity' by mean , since no skewness
df_train['Popularity'] = df_train['Popularity'].fillna(df_train['Popularity'].mean())
df_test['Popularity'] = df_test['Popularity'].fillna(df_train['Popularity'].mean())
```

```python
df_train['key'].value_counts()
```

```python
mode_val = df_train['key'].mode()[0]
df_train['key_missing'] = df_train['key'].isnull().astype(int)
df_train['key'] = df_train['key'].fillna(mode_val)
df_test['key_missing'] = df_test['key'].isnull().astype(int)
df_test['key'] = df_test['key'].fillna(mode_val)
```

```python
df_train['key'].value_counts()
```

```python
df_train.head(10)
```

```python
median_instrmnt = df_train['instrumentalness'].median()
df_train['instrumentalness_missing'] = df_train['instrumentalness'].isnull().astype(int)
df_train['instrumentalness'].fillna(median_instrmnt,inplace=True)
df_test['instrumentalness_missing'] = df_test['instrumentalness'].isnull().astype(int)
df_test['instrumentalness'].fillna(median_instrmnt,inplace=True)

```

```python
df_train.isnull().sum()
```

```python
def reorder_columns(df):
    cols = df.columns.tolist()
    # Move 'key_missing' next to 'key'
    if 'key_missing' in cols and 'key' in cols:
        key_missing_idx = cols.pop(cols.index('key_missing'))
        key_idx = cols.index('key')
        cols.insert(key_idx + 1, 'key_missing')

    # Move 'instrumentalness_missing' next to 'instrumentalness'
    if 'instrumentalness_missing' in cols and 'instrumentalness' in cols:
        instrumentalness_missing_idx = cols.pop(cols.index('instrumentalness_missing'))
        instrumentalness_idx = cols.index('instrumentalness')
        cols.insert(instrumentalness_idx + 1, 'instrumentalness_missing')

    return df[cols]
```

```python
df_train = reorder_columns(df_train)
df_test = reorder_columns(df_test)
df_train.head()
```

```python
def standardize_duration(x):
    if pd.isnull(x):
        return x
    if x < 1000:
        return x * 60
    else:
        return x / 1000
```

```python
df_train['duration_sec'] = df_train['duration_in min/ms'].apply(standardize_duration)
df_test['duration_sec'] = df_test['duration_in min/ms'].apply(standardize_duration)
```

```python
df_train = df_train.drop(['duration_in min/ms'],axis=1)
df_test = df_test.drop(['duration_in min/ms'],axis=1)
```

```python
df_train.head(10)
```

```python
sns.countplot(data=df_train,x='Class',hue='Class')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X1=df_train.drop(columns=['Class'])
y1=df_train['Class']
X1_train , x1_test , y1_train , y1_test = train_test_split(X1,y1,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X1_train = scaler.fit_transform(X1_train)
x1_test = scaler.transform(x1_test)
```

```python
X1_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.callbacks import EarlyStopping
```

```python
classifier=Sequential()
# earlier 16 , 14,12,12,11(accuracy;0.461) 3hidden
# 16,32,40,20,11(accuracy0.4644) 3hidden
```

```python
# input layer
classifier.add(Dense(units=16,activation='relu', input_dim=X1_train.shape[1]))
```

```python
# first hidden layer
classifier.add(Dense(units=32,activation='relu'))
```

```python
# second hidden layer
classifier.add(Dense(units=64,activation='relu'))
```

```python
# third hidden layer
classifier.add(Dense(units=40,activation='relu'))
```

```python
# test foruth layer
classifier.add(Dense(units=20,activation='relu'))
```

```python
# output layer
classifier.add(Dense(units=11,activation='softmax'))
```

```python
classifier.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
earlystop=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0
)
```

```python
y1_train_encoded = tf.keras.utils.to_categorical(y1_train, num_classes=11)
model = classifier.fit(X1_train, y1_train_encoded, validation_split=0.33, batch_size=50, epochs=1000, callbacks=earlystop)
```

```python
plt.plot(model.history['accuracy'])
plt.plot(model.history['val_accuracy'])
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(model.history['loss'])
plt.plot(model.history['val_loss'])
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
y1_test_encoded = tf.keras.utils.to_categorical(y1_test, num_classes=11)
loss, accuracy = classifier.evaluate(x1_test, y1_test_encoded)
print("Test Accuracy:", accuracy)
```

```python
y_pred_prob = classifier.predict(x1_test)
```

```python
y_pred1 = np.argmax(y_pred_prob, axis=1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(np.argmax(y1_test_encoded, axis=1), y_pred1))
```
